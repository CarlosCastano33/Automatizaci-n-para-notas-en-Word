```python
import pandas as pd
import datetime as datetime
from docxtpl import DocxTemplate
```

```python
doc = DocxTemplate("template.docx")
```

```python
name = "Carlos Castaño"
phoneNumber = "3118203322"
email = "carlos@gmail.com"
date = datetime.today().strftime("%d/%m/%Y")
```

```python
constants = {
    "nombre": name, "telefono": phoneNumber,
    "email": email, "fecha": date
    }
```

```python
doc.render(constants)   # Hace la modificación de los datos en el documento
doc.save("output.docx") # Guarda el documento con los datos modificados
```

```python
df = pd.read_excel("notas.xlsx")
```

```python
for index, row in df.iterrows():
    contenido: {'nombre_alumno':row['Nombre del alumno'],
    'nota_mat':row['Mat'],
    'nota:fis':row['Fis'],
    'nota_qui':row['Qui']
    }
    contenido.update(constants)
    doc.render(contenido)
    doc.save(f"Notas_de_{row['Nombre del alumno']}.docx")
```
